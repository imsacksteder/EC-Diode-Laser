```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import pandas as p
from scipy.stats import stats
from scipy.optimize import curve_fit
import scipy.optimize
```

```python
#read csv files using panda
h = p.read_csv('spectrometerDataReconfig.csv')
```

```python
#plot data
plt.plot(h.current_10460_mA,h.wavelength_10460_nm, 'bo', label='Data')

#fit a line
def test(x, m, b): 
    return m*x+b
param, param_cov = curve_fit(test, h.current_10460_mA, h.wavelength_10460_nm)

#plot fitted line
plt.plot(h.current_10460_mA, test(h.current_10460_mA, *param), 'r', label = 'Fitted line')

#labels
plt.xlabel('Current (mA)')
plt.ylabel('wavelength of max. peak (nm)')
plt.title('10460 kΩ: peak location vs. current')
plt.legend(loc='lower right')
```
