```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import pandas as p
from scipy.stats import stats
from scipy.optimize import curve_fit
import scipy.optimize
```

```python
h = p.read_csv('spectrometer_fixedcurrent_changingtemp.csv')
h
```

```python
plt.plot(h.tempC, h.wavelength_current20mA, '-r', label='20mA')
plt.plot(h.tempC, h.wavelength_current19mA, 'DarkOrange', label='19mA')
plt.plot(h.tempC, h.wavelength_current18mA, '-y', label='18mA')
plt.plot(h.tempC, h.wavelength_current17mA, '-g', label='17mA')
plt.plot(h.tempC, h.wavelength_current16mA, 'DarkGreen', label='16mA')
plt.plot(h.tempC, h.wavelength_current15mA, '-c', label='15mA')
plt.plot(h.tempC, h.wavelength_current14mA, '-b', label='14mA')
plt.plot(h.tempC, h.wavelength_current13mA, '-m', label='13mA')
plt.plot(h.tempC, h.wavelength_current12mA, 'Purple', label='12mA')

degree_sign= u'\N{DEGREE SIGN}'

plt.xlabel('Temperature ('+degree_sign+'C)')
plt.ylabel('Wavelength (nm)')
plt.title("Wavelength vs. Temperature at fixed Current")
plt.legend(loc='lower right')
```

```python
plt.plot(h.tempC, h.wavelength_current12mA, 'bo', label='Data Points')

#fit a line
def test(x, m, b): 
    return m*x+b
param, param_cov = curve_fit(test, h.tempC, h.wavelength_current12mA)

#plot fitted line
plt.plot(h.tempC, test(h.tempC, *param), 'r', label = 'Fitted line')

degree_sign= u'\N{DEGREE SIGN}'


plt.xlabel('Temperature ('+degree_sign+'C)')
plt.ylabel('Wavelength (nm)')
plt.title("Wavelength vs. Temperature at 12 mA current")
plt.legend(loc='lower right')
```
